## Genre tally from all logged titles

```python
# %load genre_count.py
import pandas as pd

df = pd.read_csv('movies_with_genres.csv')

genre_counts = df['genre'].str.split(', ').explode().value_counts()
print(genre_counts.to_string())
```

## Filter titles by year of release

```python
import pandas as pd

df = pd.read_csv('data/watched.csv')
filtered = df[(df['Year'] >= 1999) & (df['Year'] <= 2001)][['Name', 'Year', 'Letterboxd URI']]
print(filtered.sort_values('Year').to_string())
```
